1. What is Simple Linear Regression?
   
     Simple Linear Regression is a technique that helps us predict the value of one variable (Y) based on the value of another variable (X) using a linear equation.

      **Formula= Y=a+bX+e**

      Y → Dependent variable (the one we want to predict)

      X → Independent variable (the predictor)

      a → Intercept (value of Y when X = 0)

      b → Slope (rate of change of Y with respect to X)

      e → Error term (difference between actual and predicted values)

2. What are the key assumptions of Simple Linear Regression?
    1. Linearity:
          The relationship between the independent variable (X) and dependent variable (Y) should be linear.
    2. Independence of Errors:
          The residuals (errors) — difference between actual and predicted Y — should be independent of each other.
    3. Homoscedasticity
          The variance of errors should be constant for all values of X.
    4. Normality of Errors
          The residuals (errors) should be normally distributed (bell-shaped curve).
    5. No (or minimal) Multicollinearity
          In simple linear regression, there’s only one X — so this assumption doesn’t usually apply.

3: What is heteroscedasticity, and why is it important to address in regression
models?

  Heteroscedasticity occurs when the variance of the errors (residuals) is not constant across all levels of the independent variable (X).

  **why is it important: **
  
  1. Regression coefficients (a, b) remain unbiased,
  but their standard errors become unreliable.

  2. As a result:

      t-tests and F-tests can give wrong p-values

     Confidence intervals and predictions become inaccurate

  3. This can lead to false conclusions (thinking a variable is significant when it’s not).

4. What is Multiple Linear Regression?

  Multiple Linear Regression is a statistical method used to predict the value of one dependent variable (Y) using two or more independent variables (X₁, X₂, X₃, …).

   **Formula= Y=a+b1​X1​+b2​X2​+b3​X3​+⋯+bn​Xn​+e**

   Y → Dependent variable (the one we want to predict)

  X₁, X₂, X₃...Xₙ → Independent variables (predictors)

  a → Intercept (value of Y when all X = 0)

  b₁, b₂, b₃...bₙ → Coefficients (show how much Y changes with one unit change in each X)

  e → Error term (difference between actual and predicted Y)

5. What is polynomial regression, and how does it differ from linear
regression?

Polynomial Regression is a type of regression that models the relationship between the independent variable (X) and the dependent variable (Y) as an nth-degree polynomial.

      **Formula= Y=a+b1​X+b2​X2+b3​X3+⋯+bn​Xn+e**

  Different between polynomial regression:

  In linear regression, Y changes at a constant rate with X.

6. Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)

print("Intercept (a):", model.intercept_)
print("Slope (b):", model.coef_[0])

plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

7. Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]


```python

import numpy as np
from sklearn.linear_model import LinearRegression

Area = np.array([1200, 1500, 1800, 2000])
Rooms = np.array([2, 3, 3, 4])
Price = np.array([250000, 300000, 320000, 370000])

X = np.column_stack((Area, Rooms))
Y = Price

model = LinearRegression()
model.fit(X, Y)

print("Intercept (a):", model.intercept_)
print("Coefficients (b1 for Area, b2 for Rooms):", model.coef_)

Y_pred = model.predict(X)

print("\nActual Prices:   ", Y)
print("Predicted Prices:", np.round(Y_pred, 2))

```

8. Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]



```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, Y)

Y_pred = model.predict(X_poly)

print("Intercept (a):", model.intercept_)
print("Coefficients (b1, b2):", model.coef_)

plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Polynomial Fit (Degree 2)')
plt.title('Polynomial Regression (Quadratic)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

9. Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]


```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)


residuals = Y - Y_pred

print("Intercept (a):", model.intercept_)
print("Slope (b):", model.coef_[0])
print("Residuals:", np.round(residuals, 2))

plt.scatter(X, residuals, color='purple')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals Plot')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Residuals (Y - Ŷ)')
plt.show()
```

10. Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Exp

Goal: Predict house prices (Price) using features:

Area

Number of rooms

Location

Problem:

Heteroscedasticity → non-constant variance of residuals

Multicollinearity → predictors are highly correlated


1. **Detect Heteroscedasticity**

  Plot residuals vs predicted values

  Funnel or patterned spread indicates heteroscedasticity.

  Statistical tests:

  Breusch–Pagan test

  White test

2. **Detect Multicollinearity**

Correlation matrix → check correlations between predictors

Variance Inflation Factor (VIF) → VIF > 5 or 10 indicates strong multicollinearity
